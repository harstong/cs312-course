### 3.1 - Open the temple image

- Open the given Temple image "sl_temple.jpg" and display in this notebook

```python
img = cv2.imread('sl_temple.jpg')

# In order to display using MatPlotLib, RGB channels need to be swapped
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('Sample Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

### 3.6 - Invert

- Convert/load the temple image grayscale
- Copy the temple image and invert all of the pixels (0 -> 255, 10 -> 245, 255 -> 0, etc...)
- Display this new image

```python
img = cv2.imread('s1_temple.jpg')
h, w = img.shape
print(h, w)
print(img[23][34])

x = np.array([1,2,3])
print(100 - x)

for r in range(h):
    for c in range(w):
        img[r][c] = 255 - img[r][c]
plt.imshow(img, cmap='gray')
plt.title('Sample Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

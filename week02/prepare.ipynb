The number of pixels for an image can be mearsured using DPI [Dot-per-inch](https://en.wikipedia.org/wiki/Dots_per_inch) or PPI (Pixels Per Inch).

**imread()**
- The function imread() is used to load an image using OpenCV.  There is a small example [here](https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_gui/py_image_display/py_image_display.html).

Example of loading an image as a gray scale

````python
import numpy as np
import cv2

img = cv2.imread('rgbg.png', 0)

# Display the image
plt.imshow(img, cmap='gray')
plt.title('Sample Image'), plt.xticks([]), plt.yticks([])
plt.show()

# Load an color image in grayscale
print(img)
````

````python
# Example of loading an image as a RGB image
import numpy as np
import cv2

# Load an color image in RGB
img = cv2.imread('rgbg.png', cv2.IMREAD_COLOR)
print(img)
````

Format of an image

- Images loaded with imread() are numpy arrays.
- Each color of an image is called a channel.  RGB images have 3 channels, a grayscale has one channel.
- If the image has transparency, then the image will have an alpha channel.

````python
# shape function
# To display information about an image, use the shape function
# It returns a tuple of number of rows, columns and channels
print(img.shape)

height, width, channels = img.shape
print(height, width, channels)
````

````python
# size function
# To get the size of an image in bytes, use size.
print(img.size)
````

````python
# It is very important to know the data type of a numpy so you can use it correctly.
# The function is dtype.  Review numpy functions to convert 
print(img.dtype)
````

Splitting and merging image channels

- the different colors of an image are called channels.  These channels can be split apart and merged together

````python
# You can split the image channels using the command split.
img = cv2.imread('rgbg.png')
plt.imshow(img, cmap='gray')
plt.title('Sample Image'), plt.xticks([]), plt.yticks([])
plt.show()

b, g, r = cv2.split(img)
print(r)
````

````python
print(g)
````

````python
print(b)
````

````python
# You can merge channels using merge()
newimg = cv2.merge([b, g, r])
print(newimg)
````

````python
# You can use numpy to extract the channel which is faster
b = img[:,:,0]
g = img[:,:,1]
r = img[:,:,2]
print(b)
````

````python
# Changing all reds in an image to zero
img = cv2.imread('rgbg.png', cv2.IMREAD_COLOR)
r = img[:,:,2]
print(r)

img[:,:,2] = 0

r = img[:,:,2]
print(r)
````

````python
img = cv2.imread('image1.jpg')

# Changing all reds in an image to zero
img[:,:,2] = 0

img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title('Sample Image'), plt.xticks([]), plt.yticks([])
plt.show()
````
